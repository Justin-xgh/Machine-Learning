```python
import copy, math 
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
np.set_printoptions(precision=2) #降低数组的显示精度
##创建训练样本
X_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]]) #3行4列
y_train = np.array([460, 232, 178]) #1行3列
##输出训练样本
print('X.shape= ',X_train.shape,',  X Type= ',type(X_train),'')
print(X_train)
print(f'y shape = {y_train.shape}, y type= {type(y_train)}')
print(f'{y_train}')
##构造初始w，b
b_init = 785.1811367994083
w_init = np.array([ 0.39133535, 18.75376741, -53.36032453, -26.42131618])
# print(f"w_init shape: {w_init.shape}, b_init type: {type(b_init)}")
print("w_init shape: ",w_init.shape,",  b_init type: ",type(b_init),"")
##定义单一元素循环预测函数
def predict_single_loop(x,w,b):
    """
    single predict using linear regression
    Args:
        x (ndarray): Shape (n,) example with multiple features
        w (ndarray): Shape (n,) model parameters
        b (scalar) : model parameter
    Retruns:
        p (scalar) ：prediction
    """
    n=x.shape[0]
    p=0
    for i in range(n):
        p_i=x[i]*w[i]
        p += p_i
    p = p + b
    return p
#从训练样本中取出一行进行测试
x_vec= X_train[0,:] #取第一行元素作为1-D向量
print('x_vec value=',x_vec,' x_vec shape = ',x_vec.shape,' ')
f_wb = predict_single_loop(x_vec, w_init, b_init) #带入函数进行预测
print(' f_wb value = ',f_wb,', f_wb shape = ',f_wb.shape,'')
##利用点积运算定义函数，并作上述同样步骤
def predict_dot(x,w,b):
    p = np.dot(x,w) + b
    return p
x_vec = X_train[0]
print(f"x_vec shape {x_vec.shape}, x_vec value: {x_vec}")
f_wb = predict_dot(x_vec,w_init, b_init)
print(f"f_wb shape {f_wb.shape}, prediction: {f_wb}")
##采用for循环定义多变量代价函数
def compute_cost(X, y, w, b):
    """
    Compute the cost
    Args:
        X(ndarray(m,n)) : Data, m example with n feature
        y(ndarray(m,))  : target value
        w(scalar)       : model parameter
        b(scalar)       : model parameter
    Returns:
        cost(scalar)    : cost
    """
    m=X.shape[0]
    cost= 0 
    for i in range(m):
        f_wb_i = np.dot(X[i],w) + b #线性预测函数
        cost += (f_wb_i-y[i])**2
    cost = cost/(2*m)
    return cost
cost = compute_cost(X_train, y_train, w_init, b_init) #用预先的最佳参数计算代价
print('Cost at optimal w : ',cost,'')
##最佳参数下，多变量的梯度函数
def compute_gradient (X, y, w, b):
    """
    Compute the gradient for linear regression
    Args:
        X (ndarray(m,n)) : Data, m example with n feature
        y (ndarray(m,))  : target values
        w (scalar)       : model parameter
        b (scalar)       : model parameter
    Returns:
        dJ_dw (ndarray(n,)): The gradient of the cost w.r.t the w
        dJ_db (scalar)      : The gradient of the cost w.r.t the b
        """
    m,n=X.shape  #m为行数，n为列数
    dJ_dw=np.zeros((n,)) #构建零元素的一维列向量，分配内存
    dJ_db=0
    for i in range(m):   #求和m个
        err = (np.dot(X[i], w) + b) - y[i]
        for j in range (n):   #对w中个元素求编导
            dJ_dw[j] += err*X[i,j]
        dJ_db += err
    dJ_dw = dJ_dw/m
    dJ_db = dJ_db/m
    return dJ_dw, dJ_db
tmp_dJ_dw, tmp_dJ_db = compute_gradient(X_train, y_train, w_init, b_init) #计算并且展示梯度
print('dJ_dw at initial w, b : ',tmp_dJ_dw,'')
print('dJ_db at initila w, b : ',tmp_dJ_db,'')
##定义梯度下降函数
def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters):
    """
    Performs batch(批量) gradient descent to learn thea. Updates theta by taking num_iters gradient steps with learning rate alpha
    
    Args:
        X (ndarray(m,n)) : Data, mexample with n feature
        y (ndarray(m,))  : target values
        w_in(ndarray(n,)): initial model parameter
        b_in(scalar)     : initial model parameter
        cost_function    : function to compute cost
        gradient_function: function to compute gradient
        alpha (float)    : Learning rate
        num_iters (int)  : number of iterations to run gradient descent
        
    Returns:
        w (ndarray(n,))  : Updated values of parameters
        b (scalar)       : Updated values of parameter
        """
    J_history = [] #创建一个数组来存储代价J和每次迭代后的w，方便以后作图
    w = copy.deepcopy(w_in)    #避免在函数中修改全局的w
    b = b_in
    
    for i in range (num_iters):  #计算梯度和更新参数
        dj_dw, dj_db = gradient_function(X, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
        if i<100000:   #保存每次迭代的代价J
            J_history.append(cost_function(X, y, w, b))         
                                    
        if i% math.ceil(num_iters / 10) == 0:  #每当迭代次数能被100整除就输出一次成本，如果迭代小于100就打印相同次数的迭代
            print(f'Interation {i:4d} : Cost : {J_history[-1]: 0.2f}')
                             
    return w, b, J_history  #返回最终的w，b 和 历史J用来画图
                        
initial_w = np.zeros_like(w_init)   #初始化参数，并设定迭代次数和学习速率
initial_b = 0.                            
iterations = 1000
alpha = 5.0e-7                             
w_final, b_final, J_hist= gradient_descent (X_train, y_train, initial_w, initial_b,
                                            compute_cost, compute_gradient, alpha, iterations)    #运行梯度下降算法，并输出最终的w，b
print(f'w, b found by gradient descent : {w_final}, {b_final:0.2f} ')
m,_ = X_train.shape
for i in range (m):
     print(f'prediction: {np.dot(X_train[i], w_final) + b_final:0.2f}, target value: {y_train[i]}')    #将训练数据进行预测运算，对比预测值和目标值
##绘制代价和迭代的关系图
fig,(ax1,ax2) = plt.subplots(1,2, constrained_layout=True, figsize=(12,4))
ax1.plot(J_hist)
ax2.plot(100 + np.arange(len(J_hist[100:])), J_hist[100:])
ax1.set_title("Cost vs. iteration");  ax2.set_title("Cost vs. iteration (tail)")
ax1.set_ylabel('Cost')             ;  ax2.set_ylabel('Cost') 
ax1.set_xlabel('iteration step')   ;  ax2.set_xlabel('iteration step') 
plt.show()
```
